```python
# Import Libaries
import pandas as pd
from pathlib import Path
```

```python
# File to Load 
employees_path = Path("../data/employees.csv")
departments_path = Path("../data/departments.csv")
dept_emp_path = Path("../data/dept_emp.csv")
dept_mgr_path = Path("../data/dept_manager.csv")
salaries_path = Path("../data/salaries.csv")
titles_path = Path("../data/titles.csv")
```

```python
# Read Data File and store into Pandas DataFrames
employees_data = pd.read_csv(employees_path)
departments_data = pd.read_csv(departments_path)
dept_emp_data = pd.read_csv(dept_emp_path)
dept_mgr_data = pd.read_csv(dept_mgr_path)
salaries_data = pd.read_csv(salaries_path)
titles_data = pd.read_csv(titles_path)
```

```python
#List the employee number, last name, first name, sex, and salary of each employee
employee_salaries = pd.merge(employees_data, salaries_data, how="left", on=["emp_no", "emp_no"])
employee_salaries[["emp_no","last_name","first_name","sex","salary"]]
```

```python
# List the first name, last name, and hire date for the employees who were hired in 1986
employees_hired_in_1986 = employee_salaries[employee_salaries['hire_date'].str.contains('1986', na=False)]
employees_hired_in_1986[['first_name', 'last_name', 'hire_date']]
```

```python
# List the manager of each department along with their 
#department number, department name, employee number, last name, and first name.

dept_mgr = pd.merge(dept_mgr_data, departments_data, how="left", on=["dept_no", "dept_no"])
dept_mgr_details = pd.merge(dept_mgr, employees_data, how="left", on=["emp_no", "emp_no"])
dept_mgr_details[["dept_no","dept_name","emp_no","last_name","first_name"]]
```

```python
# List the department number for each employee along with that employee’s 
# employee number, last name, first name, and department name
employee_dept = pd.merge(employees_data, dept_emp_data, how="left", on=["emp_no", "emp_no"])
employee_dept_name = pd.merge(employee_dept, departments_data, how="left", on=["dept_no", "dept_no"])
employee_dept_name[["emp_no","last_name","first_name","dept_name"]]
```

```python
# List the first name, last name, and sex of each employee whose 
# first name is Hercules and whose last name begins with the letter B
hercules_employees = employees_data[employees_data['first_name'] == 'Hercules']
hercules_b_employees = hercules_employees[hercules_employees['last_name'].str.startswith('B')]
hercules_b_employees[["first_name","last_name","sex"]]
```

```python
# List each employee in the Sales department, including their employee number, last name, and first name.
employee_dept = pd.merge(employees_data, dept_emp_data, how="left", on=["emp_no", "emp_no"])
employee_dept_name = pd.merge(employee_dept, departments_data, how="left", on=["dept_no", "dept_no"])
sales_employee = employee_dept_name[employee_dept_name["dept_name"] == "Sales"]
sales_employee[["emp_no","last_name","first_name"]]
```

```python
# List each employee in the Sales and Development departments, 
# including their employee number, last name, first name, and department name
employee_dept = pd.merge(employees_data, dept_emp_data, how="left", on=["emp_no", "emp_no"])
employee_dept_name = pd.merge(employee_dept, departments_data, how="left", on=["dept_no", "dept_no"])
sales_dev_employee = employee_dept_name[(employee_dept_name["dept_name"] == "Sales") | (employee_dept_name["dept_name"] == "Development")]
sales_dev_employee[["emp_no","last_name","first_name"]]
```

```python
# List the frequency counts, in descending order, of all the employee last names 
# (that is, how many employees share each last name).
last_name_counts = employees_data['last_name'].value_counts()
last_name = pd.DataFrame(last_name_counts)
last_name.sort_values(by='count', ascending=False)
```
